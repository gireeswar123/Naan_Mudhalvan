***Load the dataset.***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv('/content/drive/MyDrive/HousePriceIndia.csv')
df.head()


```

Perform the Visualizations.***Univariate Analysis***

```python
sns.distplot(df.Price)
```

```python
sns.displot(df.Lattitude)
```

```python
sns.barplot(x=df.Built_Year.value_counts().index,y=df.Built_Year.value_counts())
```

```python
sns.barplot(x=df.Number_of_schools_nearby.value_counts().index,y=df.Number_of_schools_nearby.value_counts())
```

```python
plt.pie(df.Number_of_schools_nearby.value_counts(),[0.1, 0, 0],labels=[1,2,3],autopct="%1.1f%%",colors=['red','yellow','blue'],shadow=True)
plt.title('Number_of_schools_nearby')
plt.show()
```
